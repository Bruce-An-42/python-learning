```python
hrs = input ("Enter hours:")
h = float (hrs)
rate = input ("Enter rate:")
r = float (rate)
if h <= 40:
    p = h * r
elif h > 40:
    p = 40*r+(h-40)*1.5*r
print("pay=",p)
```

```python
score = input("Enter Score: ")
s = float(score)
if 0.9<=s<=1.0:
    print("A")
elif 0.8<=s<0.9:
    print("B")
elif 0.7<=s<0.8:
    print("C")
elif 0.6<=s<0.7:
    print("D")
elif s<0.6:
    print("E")

```
