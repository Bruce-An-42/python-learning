```python
def average(numbers):
    """return the arithmetic mean of a list of numbers."""
    if len(numbers) == 0:
        return None
    total = 0
    for n in numbers:
        total = total + n
    return total / len(numbers)

prices_week_1 = [100, 102, 101, 103, 105]
prices_week_2 = [105, 104, 106, 107, 108]
print(f"week 1 average: {average(prices_week_1):.2f}")
print(f"week 2 average: {average(prices_week_2):.2f}")
```
